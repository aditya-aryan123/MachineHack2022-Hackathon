```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
df_train = pd.read_csv('train.csv')
df_train.head()
```

```python
df_train.shape
```

```python
df_train_copy = pd.get_dummies(data=df_train, columns=['AGE', 'DRIVING_EXPERIENCE', 'TYPE_OF_VEHICLE'])
```

```python
le = LabelEncoder()
df_train_copy['GENDER']=le.fit_transform(df_train['GENDER'])
df_train_copy['EDUCATION']=le.fit_transform(df_train['EDUCATION'])
df_train_copy['INCOME']=le.fit_transform(df_train['INCOME'])
df_train_copy['VEHICLE_YEAR']=le.fit_transform(df_train['VEHICLE_YEAR'])
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit_transform(df_train_copy)
```

```python
inertia = []
range_val = range(1,15)
for i in range_val:
    kmean = KMeans(n_clusters=i)
    kmean.fit_predict(pd.DataFrame(scaled_df))
    inertia.append(kmean.inertia_)
plt.plot(range_val,inertia,'bx-')
plt.xlabel('Values of K') 
plt.ylabel('Inertia') 
plt.title('The Elbow Method using Inertia') 
plt.show()
```

```python
kmeans_model=KMeans(5)
kmeans_model.fit_predict(scaled_df)
df_kmeans= pd.concat([df_train, pd.DataFrame({'cluster': kmeans_model.labels_})],axis=1)
```
